#### 案例: Audio Categorization
http://lpis.csd.auth.gr/publications/tsoumakas-ismir08.pdf

##### 目標
因為音樂可以喚起多種情感的關聯

將支持實現通過情感查詢音樂收藏的音樂信息檢索系統

個人感覺是可以把情感做一種量化指標

##### 資料來源
7種不同流派的每首100首歌曲：古典，雷鬼，搖滾，流行，嘻哈，電子舞曲和爵士樂。

從每首歌曲中提取最初的30秒後的30秒。 存儲所得的聲音片段，並將其轉換為22050 Hz採樣率，每個樣本16位和單聲道的波形文件

提取的特徵分為兩類：節奏和音色

- 節奏特徵

通過從拍子直方圖中提取周期性變化而得出的。 實現了使用自相關識別峰的算法。

選擇了兩個最高峰，併計算了它們的振幅，各別的BMP（每分鐘幾拍）和各別的BPM的高低比。

此外，通過分別將40-90、90-140和140-250 BPM之間的直方圖塊加起來，計算出3個特徵。 整個過程共有8種節奏特徵。

--------------------

- 音色特徵

使用 Mel Frequency Cepstral Coefficients (MFCCs) 建模

將信號分為多個幀(frame)，並為每個幀計算幅度頻譜。再將其轉成 Mel scale 

選用前13個MFCCs來使用。

再利用音色透過 Short-Term Fourier Transform (FFT)得出3個特徵:

Spectral centroid, spectral rolloff and spectral flux.

- 小結

節奏有8個特徵: 兩個高峰合計振幅，各別的BMP，各別的高低比(5個)+3段BPM的累計 (3個)

音色有16個特徵（13個MFCC，3個FFT）


##### Emotion Labeling

只有6種感情；由作者大學音樂學院的三名年齡分別為20、25和30的男性專家註釋。

僅保留所有專家標記完全相同的歌曲。剩593首



感情根據Tellegen-Watson-Clark model

#### 方法

使用Multilabel Classification效果最好 (比較Single Classification 和 回歸)

而其中核心以RAKEL最為出色

- 比較:
    - binary relevance (BR)
    - label powerset (LP)
    - randomk-labelsets (RAKEL)
    - multilabel k-nearest neighbor (MLkNN)
